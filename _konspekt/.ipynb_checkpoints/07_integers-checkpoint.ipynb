## Целочисленные типы

в С++ кроме int есть и другие целочисленные типы, про которые нужно знать. 

один из самых распространенных это unsigned int, который имеет такой же размер как и int, но не может хранить отрицательные значения

например, если у нас есть вектор vector <int> t = {0, 1, 2};, который является целочисленным int вектором. При работе с ним могут возникнуть проблемы, если мы забудем о некоторых свойствах целочисленных типов. например t.size() - возвращает размер вектора, который является типом unsigned int, т.е. размер вектора не может быть отрицательным
Чтобы избежать проблем с ним, мы можем использовать 
```
static_cast<int> t.size()
```
что конвертирует unsigned тип в обычный int


    
Рассмотрим реализацию цикла for в этом случае:
```
for (int i = 0; i< static_cast<int> t.size(); ++i) {
    ...
}
```
или 
```
for (size_t i = 0; i< t.size(); ++i){
    ...
}
```

Во втором случае мы объявили i типа unsigned, поэтому переменная не может уйти в -

Следует, также, помнить о том, что при сложении, сравнении, умножении двух переменных, компилятор с++ автоматически конвертирует их в больший из двух типов.

Пример:

* Если у нас есть int lhs и long int rhs, то при lhs + rhs, в ответе будет тип long int. 

* Если у нас unsigned lhs и int rhs, то при любой операции с этими двумя переменными ответ будет сконвертирован в unsigned 

**Следует помнить, что размер типа int зависит от системы в которой компилируется программа, в 64 разрядной системе это 4 байта, в 32 разрядной 2 байта**
