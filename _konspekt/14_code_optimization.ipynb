### Быстрый код

что такое быстрый код?

* код не бывает просто быстрым или медленным

* он бывает достаточно быстрым для своей задачи

Отсюда вывод:

**Избегайте преждевременной оптимизации**

Для большинства задач действует **Принцип Парето** (как и для многих задач в жизни, кстати говоря)

* 80% времени работы программы тратится на исполнение 20% кода

* эти 20% и надо оптимизровать

* Но сначала их надо найти

### Недостатки преждевременной оптимизации

* необоснованное усложнение кода

* усложнение поддержки кода

* замедление разработки

### Итоги

* Как правило, код надо стараться делать правильным, простым, а только потом быстрым

* Это не значит, что о производительности вообще не надо думать

### 2 правило оптимизации

Прежде, чем ускорять код, замертье сколько он работает

Если это недостаточно быстро, начинайте искать узкие места

Интуиция не работает --- замеряйте

### Разница между endl и "\n"

Наиболее наглядным примером разницы между endl (перевод на следующую строку) и '\n' (знак перевода на следующую строку) является следующий код:
```
for (int i = 0; i < 10000; ++i){
    cout << i << '\n';
}
```
и 
```
for (int i = 0; i < 10000; ++i){
    cout << i << endl;
}```

Если замерить время работы этих кусков программы, то мы заметим, что второй код, с endl будет работать медленнее.

Это проихсодит потому, что cout отправляет в поток, который хранится в буфере. 

* endl сразу сбрасывает буфер, 

* "\n" не сбрасывает в буфер. Т.е. буфер сбрасывается при переполнении

Еще одним способов сброса буфера является использование **cin** после **cout**, т.к. и ввод в поток, и вывод из потока используют один и тот же поток. Поэтому перед использованием потока, cin сбрасывает буфер

Можно этого исбежать использовав команду:
```
cin.tie(nullptr);
```

### Сложность алгоритмов

**Линейная (N):**

* обычный цикл for (с легкими итерациями)

* алгоритмы find_if, min_element и пр.

**NLog(N)**:

* алгоритм sort (с простыми сравнениями)

часто используют O-символику:

О(N), O(NLogN) и т.д.

Это означает, что алгоритм работает не больше, чем N или NLogN. O - это оценка сверху (она может никогда не достигаться(
