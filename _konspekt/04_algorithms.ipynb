## Стандартные алгоритмы

В стандартной библиотеке C++ имеются наиболее часто используемые алгоритмы, такие как: мин, макс, сортировка, подсчет совпадающих элементов и т.д.

Стандартная библиотека алгоритмов вызывается командой 
```
#include <algorithm>
```

Ниже приводятся некоторые команды из стандартной библиотеки алгоритмов:

**count**(begin, end, elem) --- количество элементов в интервале begin и end.

**count**(begin, end, pred) --- количество элементов удовлетворяющих условию pred.

в качестве условия pred можно использовать функцию, либо лямбда-функцию:
```
[t](int x) {return x>t;}
```
Данная лямбда-функция возвращает true если элемент больше t, и false если меньше

Лямбда-функция начинается с квадратной скобки [], внутри которой можно добавлять переменные, которые будут передаваться с другой части программы (в которой находится лямбда-функция), после чего идут круглые скобки () внутри которой пишутся переменные, которые должны быть переданы при вызове функции. в фигурных скобках {} приводится само тело программы.

Ниже приводится небольшой пример программы:

```
vector<int> v = {1, 3, 2, 4, 5};
cout << count(begin(v), end(v), 2);
```
будет выводить, сколько раз встречается число 2 в векторе.

Другой пример, с лямбда функцией:
```
vector<int> v = {1, 3, 2, 4, 5};
t = 2;
cout << count_if(begin(v), end(v), [t](int x) {return x>t;});
```
будет выводить, сколько элементов больше t=2.
